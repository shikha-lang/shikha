```python
#Importing orders_products_merged_busiest.pkl dataframe

orders_products_merged = pd.read_pickle(os.path.join(path, '02-Data','Prepared-Data','orders_products_merged_busiest.pkl'))
```

```python
# create subset
df = orders_products_merged[:1000000]
```

```python
# apply groupby function
df.groupby('product_name')
```

```python
#  Aggregation of department _id and order_number coloum then calculate the mean for each "order_number" column
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Group by of department_id and mean of order_number
df.groupby('department_id')['order_number'].mean()
```

```python
#Aggregation of department _id and order_number coloum then calculate the mean,min and max value for each "order_number" column
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
#Group data by "user_id" column, generate max order for each user, create "max_order" for aggregation results
orders_products_merged['max_order'] = orders_products_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# head of orders_products_merged dataframe
orders_products_merged.head(15)
```

```python
orders_products_merged.head(100)
```

```python
pd.options.display.max_rows = None
```

```python
# creating deriving coloum with log function
orders_products_merged.loc[orders_products_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
orders_products_merged.loc[(orders_products_merged['max_order'] <= 40) & (orders_products_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
orders_products_merged.loc[orders_products_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#  checking head of orders_products_merged dataframe
orders_products_merged.head()
```

```python
# Check frequencies of loyalty flags for accuracy
orders_products_merged['loyalty_flag'].value_counts(dropna=False)
```

# Question 6

```python
# apply groupby function
df.groupby('prices')
```

```python
# Split the data into groups by "loyalty_flag" and then calculate the mean for each "prices" column
orders_products_merged.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
# Group data by "user_id" column, generate mean product price for each user, create "mean_price" for aggregation results
orders_products_merged['mean_price'] = orders_products_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check grouping
orders_products_merged.head(100)
```

```python
#creating deriving coloum with log

orders_products_merged.loc[orders_products_merged['mean_price'] >= 10, 'spending_flag'] = 'High Spender'
orders_products_merged.loc[(orders_products_merged['mean_price'] < 10) , 'spending_flag'] = 'Low Spender'
```

```python
# Check frequencies of spending flags for accuracy
orders_products_merged['spending_flag'].value_counts(dropna=False)
```

# Question 7

```python
# Group data by "user_id" column, generate median product of "days since prior order, create "median_days" for aggregation results
orders_products_merged['median_days'] = orders_products_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check grouping
orders_products_merged.head(100)
```

```python
#creating deriving coloum with log

orders_products_merged.loc[orders_products_merged['median_days'] >=20, 'frequency_flag'] = 'Non-frequent customer'
orders_products_merged.loc[(orders_products_merged['median_days'] > 10)&(orders_products_merged['median_days']<=20),'frequency_flag'] = 'Regular customer'
orders_products_merged.loc[orders_products_merged['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequencies of spending flags for accuracy
orders_products_merged['frequency_flag'].value_counts(dropna=False)
```
